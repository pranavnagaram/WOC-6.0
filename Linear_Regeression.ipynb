##**Implementation:**
<ul>
<li>Created a Class LinReg</li>
<li>Defined learning rate(alpha) and iterations using init method</li>
<li>Defined a Normalize function using the below formula</li>
<li>Defined a Train function for training the model using the below listed formulae</li>
<li>Defined Predict function to make prediction and calculate error</li>

$$\hat{y} = \mathbf{x} \cdot \mathbf{w} + b$$
###Normalization:
$$x_i = \frac{{x_i - \mu}}{{\sigma}}$$
###Cost Function:

$$J = \frac{1}{2m} \sum_{i=1}^{m} (\hat{y}_i - y_i)^2  \implies J = \frac{1}{2m} \sum_{i=1}^{m} (\mathbf{x}_i \cdot \mathbf{w}_i - b)^2$$
###Gradient:
$$\frac{\partial J}{\partial w_i}= \frac{1}{m} \sum_{i=1}^{m} (\mathbf{x}_i \cdot \mathbf{w}_i - b)\cdot\ \mathbf{x}_i$$

$$\frac{\partial J}{\partial b}= \frac{1}{m} \sum_{i=1}^{m} (\mathbf{x}_i \cdot \mathbf{w}_i - b)$$
###Gradient Descent:
$$w_i = w_i - \alpha \frac{\partial J}{\partial w_i}$$

$$b = b - \alpha \frac{\partial J}{\partial b}$$












```python
class LinReg:
   def __init__(self,alpha,iterations):
       self.alpha=alpha
       self.iterations=iterations
   def Normalize(self,x):
       self.sigma=np.std(x,axis=0)
       self.mu=np.mean(x,axis=0)
       X=(x-self.mu)/self.sigma  ##z-normalization
       return X
   def Train(self,x,y):
       X=self.Normalize(x)
       m=X.shape[0]         ##number of samples
       n=X.shape[1]         ##number of features
       self.w=np.random.randn(n,1)*np.sqrt(2/n)  #He initialization
       self.b=np.ones((1,1))
       self.J_change=[]
       for i in range(self.iterations):
         J=(1/(2*m))*np.sum((np.dot(X,self.w)+self.b-y)**2)
         self.J_change.append(J)
         f=np.dot(X,self.w)+self.b-y
         self.w -= (1/m)*(self.alpha)*np.dot(X.T,f)
         self.b -= (1/m)*(self.alpha)*np.sum(f)
         if i%200==0:
          print(f"Cost after {i}th iteration is: {J}")
       plt.figure(figsize=(4,3))
       plt.plot(np.arange(self.iterations)[:80],self.J_change[:80],'r')
       plt.xlabel("Iterations (till 80)")
       plt.ylabel("Cost Function (J)")
       print(self.w,self.b)
   def Predict(self,x):
       X=self.Normalize(x)
       Y=np.dot(X,self.w)+self.b
       return Y
   def Error(self,x,y):
      y_p=self.Predict(x)
      e= np.mean(abs(y_p-y)/abs(y))*100
      return print(f"Error(%) is {e}%")
   def R2score(self,x,y):
       y_pred = self.Predict(x)
       rss = np.sum((y_pred-y)**2)
       tss = np.sum((y-np.mean(y))**2)
       r2 = 1 - (rss)/(tss)
       return print(f'R2_score :{r2}')
```

```python
plt.figure(figsize=(15, 13))
for i in range(x_train.shape[1]):
    plt.subplot(4,5,i+1)
    plt.plot(x_train[:100, i],label=f'Feature {i+1}',color='blue')
    plt.plot(LinReg(0.3,1000).Normalize(x_test)[:100,i],label=f'Normalized {i+1}',color='green')
    plt.xlabel('Data Points')
    plt.legend()
plt.show()
```
